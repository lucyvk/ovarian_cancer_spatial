```python
# The above TSNE values are saved in the cell table in the columns tsne_1 and tsne_2
df = pd.read_csv(os.path.join("Data","cell_table.csv"))
cell_type_colors = viz_helper.get_cell_type_colors()
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
for key, value in cell_type_colors.items():
    ax.scatter(df.loc[df.CellType == key ,'tsne_1'], df.loc[df.CellType==key , 'tsne_2'],s=0.05, c=value, label=key)
    ax.set_xlabel("tsne1", fontsize=10)
    ax.set_ylabel("tsne2", fontsize=10)
legend = plt.legend(loc="center left", borderaxespad=0, bbox_to_anchor=(1.05, 0.5))
for handler in legend.legendHandles:
    handler.set_sizes([50])
plt.savefig('Figure_2/tsne_plot.png',dpi=1000,bbox_inches='tight')
plt.show() 
```
